## Create a function in R that finds if a number is even or not.

```r
is.even <- function(a.number){
    remainder <- a.number %% 2
    if(remainder==0)
        return(TRUE)
    return(FALSE)
}
```

```r
is.even(15)
is.even(0)
```

```r
# if the function has 2 parameters that are used in the function, then you must call the function with 2 parameters; else you'll get an error

is.even2 <- function(a.number, b.number){
    remainder <- a.number %% 2
    print(b.number)
    if(remainder==0)
        return(TRUE)
    return(FALSE)
}

is.even2(15)
```

## Lambda and apply

```r
messy.vector <- c(8, 6, NA, 7, 5, NA, 3, 0, 9)
sapply(messy.vector, is.even) # error because messy.vector has NA
```

```r
# create a new is.even function that takes NA into consideration
is.even3 <- function(a.number){
    if(is.na(a.number))
        return(NA)
    remainder <- a.number %% 2
    if(remainder==0)
        return(TRUE)
    return(FALSE)
}

sapply(messy.vector, is.even3)

# NOTE: null is a value. the element is removed. its not there. its a boolean
# NA is a missing value
```

```r
# sapply goes through every element and applies a function to those values
??sapply
```

```r
# you can assign the funciton within the sapply
where.even <- sapply(messy.vector, is.even3)
where.div.3 <- sapply(messy.vector, function(num) {
    is.divisible.by(num, 3)
})
# "&" is like the "&&" and operator but for vectors
messy.vector[where.even & where.div.3]
```

```r
library(XML)

# Parse the XML content

xmlFile <- xmlTreeParse("example1 (1).xml") # parse doesnt return the actual xml document, just the contents of it
xmlValue(xmlFile$doc$children$the_beatles[["formed"]]) # gives the value between formed.
# to get a value within the tree structure ~ function is like a hierarchal tree. have to get through each layer to get to the value
xmlValue(xmlFile$doc$children$the_beatles$members$member[0][["first_name"]]) # have to identify which member you want the first name of
xmlFile

# root node: the beatles
# elements: formed, membesr, member, first_name, last_name (all tags in the XML)
```

```r
# to return the whole document by tree format
xmlFile$doc$children
xmlFile$doc$children$the_beatles
```

```r
xmlChildren(root)
```

```r
# xmlRoot gives everything under the root
root <- xmlRoot(xmlFile)
print(root)
# applying a function to the XML
# is every element in the collection
sapply(xmlChildren(root[["members"]]), function(x){
    print(xmlValue(x[["first_name"]])) # you're getting the value within first_name
})

# first x in iteration is:
#<member>
#   <first_name>George</first_name>
#   <last_name>Harrison</last_name>
#  </member

# second x in iteration is:
#<member>
#   <first_name>Ringo</first_name>
#   <last_name>Starr</last_name>
#  </member>

# etc...
```

```r
# how to get the last names
root <- xmlRoot(xmlFile)
# applying a function to the XML
# is every element in the collection
sapply(xmlChildren(root[["members"]]), function(x){
    print(xmlValue(x[["last_name"]])) # you're getting the value within first_name
})
```

```r
beatlesFile <- xmlParse("example1 (1).xml")
all_first_names = "//members/member/first_name"   # there are multiple instances of member and first name and they fall under member, which is 1 entity. you can use // to shorten the path
all_last_names = "//members/member/last_name"
xmlValue(getNodeSet(beatlesFile, all_first_names))  # getNodeSet gets all first names
xmlValue(getNodeSet(beatlesFile, all_last_names))
```

```r
create_artist_query_url_mb <- function(artist){
  encoded_artist <- URLencode(artist)  # allows you to convert URL in a way that the web can understand ~ like how spaces ' ' gets converted to a '%' in URL encoding
    return(paste0("http://musicbrainz.org/ws/2/artist/?query=artist:", encoded_artist))  # this link gives us an XML file
}

create_artist_query_url_mb('Taylor Swift')  # Taylor Swift converts to Taylor%Swift in a URL

ns <- "http://musicbrainz.org/ns/mmd-2.0#"
names(ns)[1] <- "ns"
```

```r
library(httr)
get_tag_vector_mb <- function(an_artist, ns){ # gives a vector of tags about this singer's music
    artist_url <- create_artist_query_url_mb(an_artist)
    doc <- GET(artist_url, accept("application/xml"))  # web calls (e.g. get, put, post)
    the_xml <- xmlParse(doc)  # parse through the doc
    # //ns:artist[1] because the first artist is the artist we want
    # then you can directly get the tag-list and the name from there
    # xpath <- "//ns:artist[1]/ns:tag-list/ns:tag/ns:name"
    xpath <- "//ns:tag-list/ns:tag/ns:name"
    the_nodes <- getNodeSet(the_xml, xpath, ns)
    return(unlist(lapply(the_nodes, xmlValue))) # takes the values out of the list. class will be string. if you want list then use list()
}
doc
get_tag_vector_mb('Taylor Swift', ns)
```
