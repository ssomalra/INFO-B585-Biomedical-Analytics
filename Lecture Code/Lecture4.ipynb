```r
# loading the XML library
library(XML)
# loading the XML file
patFile <- "563-ws-training.xml"
# creating a pointer to the XML file
doc <- xmlParse(patFile)
# getting <event...> Nodes from the XML using XPath
events <- getNodeSet(doc, "//patient/glucose_level/event")
# getting value of event tag and converting it into numeric
glucoseVals <- sapply(events, function(x){as.numeric(xmlAttrs(x)[["value"]])})
```

```r
glucoseVals
```

```r
# doing a basic plot
plot(glucoseVals)
```

```r
# doing a histogram
hist(glucoseVals)
```

```r
# a more fancy density histogram
plot(density(glucoseVals))
```

# Working with Time Series Data

```r
# get timestamp in the correct format
glucoseTimes <- sapply(events, function(x){as.POSIXct(xmlAttrs(x)[["ts"]], format="%d-%m-%Y %H:%M:%S")})  # ts is where the timestamps are located
# make it into printable time using as.POSIXct()
glucoseTimes <- as.POSIXct(glucoseTimes, origin = "1970-01-01")
```

```r
# try basic plot
plot(glucoseTimes, glucoseVals)
# make a data frame
glucoseData <- data.frame(glucoseTimes=glucoseTimes, glucoseVals=glucoseVals)
```

```r
glucoseData
```

```r
# make a ggplot with line
library(ggplot2)
ggplot(data=glucoseData, aes(x=glucoseTimes, y=glucoseVals)) + geom_line()
```
